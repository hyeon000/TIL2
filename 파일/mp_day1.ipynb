```python
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["C1", "C2"]
df4.index.names = ["R1", "R2"]
```

```python
df4
```

```python
#stack(열->행), unstack(행->열) 함수 : 행 인덱스 <-> 열 인덱스
```

```python
df4.stack('C1')
df4.stack('C2')
```

```python
df4
```

```python
df4.stack(0)
df4.stack(1)
```

```python
df4
```

```python
df4.unstack('R1')
```

```python
df4.unstack('R2')
```

```python
df4
```

```python
df4.swaplevel("R1","R2")
df4.swaplevel("R1","R2", 0) #default axis=0
df4.swaplevel("C1","C2", 1)
```
